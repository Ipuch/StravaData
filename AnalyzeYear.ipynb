```python
import matplotlib.pyplot as plt
from matplotlib.dates import (YEARLY, DateFormatter, rrulewrapper, RRuleLocator, drange, MonthLocator, WeekdayLocator, num2date, date2num)
import numpy as np
import datetime
```

```python
# tick every 5th easter
#rule = rrulewrapper(MONTHLY, byeaster=1, interval=1)
#loc = RRuleLocator(rule)
loc=MonthLocator(bymonth=None, bymonthday=1, interval=1)
formatter = DateFormatter('%m/%y') #'%d/
# generate dates for the year 2020
date1 = datetime.date(2020, 1, 15)
print(date1)
# generate dates for the year 2020
date2 = datetime.date(2020, 12, 15)
delta = datetime.timedelta(days=5)
# range of dates
dates = drange(date1, date2, delta)
# x min/km values random
s = np.random.rand(len(dates))*2+4  # make up some random y values

# plot the figure
fig, ax = plt.subplots()
plt.plot_date(dates, s,label='min/km')

ax.xaxis.set_major_locator(loc)
ax.xaxis.set_major_formatter(formatter)
ax.xaxis.set_tick_params(rotation=30, labelsize=10)
ax.xaxis.axis_name = 'Date'

plt.xlabel('Date')
plt.ylabel('min/km')
plt.title('Vitesse au km pour 2020')
plt.legend()

plt.show()
```
